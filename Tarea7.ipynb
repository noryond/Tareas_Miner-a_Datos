```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
data = pd.read_excel('/content/drive/My Drive/Colab Notebooks/Archivos/Proyecto1/bd_vi_2024.xlsx')
```

```python
data.head()
```

```python
data = data.dropna(subset=[
    "MEDIDAS_SEGURIDAD",
    "VIC_EDAD",
    "VIC_SEXO",
    "HEC_DEPTO",
    "AGR_SEXO",
    "HEC_TIPAGRE"
])
data["MEDIDAS_SEGURIDAD"] = pd.to_numeric(data["MEDIDAS_SEGURIDAD"], errors="coerce").fillna(2)
data.head()
```

```python
data["HEC_TIPAGRE"] = data["HEC_TIPAGRE"].apply(lambda x: 1 if x == 1 else 0)
```

```python
X = data[["VIC_EDAD", "VIC_SEXO", "HEC_DEPTO", "AGR_SEXO"]]
X = X.fillna(0)
y = data["MEDIDAS_BIN"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential([
    Dense(16, activation="relu", input_dim=X_train.shape[1]),
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")   # Para binaria
])

model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

```python
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```
